```python
import yfinance as yf


hise=yf.Ticker("AAPL")
data=hise.history(period="1y")

data.head()
```

```python
import matplotlib.pyplot as plt

data[["Close","Open"]].plot(kind="line",figsize=(40,10))
plt.show()
```
